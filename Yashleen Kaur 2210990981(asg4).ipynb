# WHAT IS UNIVARIATE ANALYSIS

# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import matplotlib.pyplot as plt
survival_counts = df['Survived'].value_counts()
plt.figure(figsize=(8, 6))
survival_counts.plot(kind='bar', color=['blue', 'red'])
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.title('Number of Survivors and Non-Survivors')
plt.xticks([0, 1], ['Died', 'Survived'], rotation=0)
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
# Count the number of passengers in each class
passenger_counts = df['Pclass'].value_counts().sort_index()

# Plotting the counts
plt.figure(figsize=(8, 6))
passenger_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers in Each Passenger Class')
plt.xticks(rotation=0)
plt.show()
```

# BIVARIANT ANALSIS

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1) 
plt.hist(df['PassengerId'], color='blue', bins=20)
plt.xlabel('PassengerId')
plt.ylabel('Frequency') 
plt.title('Histogram of PassengerId') 
plt.subplot(1, 2, 2) 
plt.hist(df['Age'], color='red', bins=20) 
plt.xlabel('Age')  
plt.ylabel('Frequency') 
plt.title('Histogram of Age')
plt.tight_layout()
plt.show()
```

# WHO HAS BETTER CHANCE OF SURVIVAL MALE OR FEMALE

```python
survival_rates_by_gender = df.groupby('Sex')['Survived'].mean()

plt.figure(figsize=(8, 6))
survival_rates_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.xticks(rotation=0)
plt.show()
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
survival_rates_by_class = df.groupby('Pclass')['Survived'].mean()
best_class = survival_rates_by_class.idxmax()
plt.figure(figsize=(8, 6))
survival_rates_by_class.plot(kind='bar', color='skyblue')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.xticks(rotation=0)
plt.show()
```
